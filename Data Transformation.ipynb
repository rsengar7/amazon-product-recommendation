```python
df = pd.read_csv("dtype_clean.csv")
df.head()
```

```python
df['category'].unique()
```

```python
category_list = []
for i in df['category'].unique():
  try:
    if "camera" in i.lower() or "television" in i.lower():
      print(i)
      print()
      category_list.append(i)
      pass
  except:
    pass
    # print(i)
```

```python
# col = ['Electronics|Camera & Photo|Printers & Scanners', '']
category_list
```

```python
data = df[df['category'].isin(category_list)]
data.head()
```

```python
df.info()
```

```python
data.info()
```

```python
drop_col = ["Unnamed: 0.1", "Unnamed: 0", "tech1","isbn_10", "isbn_13", "unspsc_code", "model_number", "style"]
data = data.drop(drop_col, axis=1)
data.info()
```

```python
data = data.fillna("")
```

```python
data_drop = data.dropna(axis=1)
data_drop.info()
```

```python
data_drop.head()
```

```python
def get_product(text):
  try:
    return text.split("|")[3]
  except:
    return ""

def get_category_type(text):
  try:
    return text.split("|")[2]
  except:
    return ""


data_drop['category_main'] = data_drop['category'].apply(lambda x: x.split("|")[0])
data_drop['sub_category'] = data_drop['category'].apply(lambda x: x.split("|")[1])
data_drop['category_type'] = data_drop['category'].apply(get_category_type)
data_drop['category_product'] = data_drop['category'].apply(get_product)

data_drop['category_main'].iloc[0]
```

```python
data_drop.head()
data_drop.category_product.unique()
```

```python
reindex = ['category', 'category_main', 'sub_category',
       'category_type', 'category_product', 'title', 'also_buy', 'brand', 'feature', 'also_view',
       'main_cat', 'price', 'asin', 'imageURL', 'imageURLHighRes', 'overall',
       'vote', 'verified', 'reviewTime', 'reviewerID', 'reviewerName',
       'reviewText', 'summary']

data_drop = data_drop[reindex]
data_drop.info()
```
